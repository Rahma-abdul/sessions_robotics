```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 3]
def displayImage(img):
    new_img= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(new_img)
    plt.show()
    
img1= cv2.imread('./pics/img1.jpg')
displayImage(img1)
img2= cv2.imread('./pics/img2.jpg')
displayImage(img2)
img3= cv2.imread('./pics/img3.jpg')
displayImage(img3)

img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))
img3 = cv2.resize(img3, (img1.shape[1], img1.shape[0]))
concatenation=cv2.hconcat([img1, img2, img3])
displayImage(concatenation)
```
